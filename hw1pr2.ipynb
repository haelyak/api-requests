```python
#
# let's wrap our requests in a function with a specific purpose
#

def get_id_from_name(artist_name):
    """ This function takes in 
           artist_name:  a string, e.g., "Lizzo" or "Taylor Swift" or ...
           and returns the artist's id#

        Return -1 to signal an error if there was no match.
    """
    # good to keep in mind: you don't _have_ to re-download data, if you have already
    # this (hand-created) table of already-found ids will save API calls (yours will differ! :-)
    if artist_name == "Kendrick Lamar": return 368183298
    
    # Not in the table above? Use the itunes search API to get an artist's itunes ID
    search_url = "https://itunes.apple.com/search"

    parameters = {  "term":artist_name,
                    "entity":"musicArtist",
                    "media":"music",
                    "limit":200  }

    result = requests.get(search_url, params=parameters)
    # print(result.url)
    # print(f"result.status_code is {result.status_code}")
    if result.status_code == 200:
        data = result.json()   # this is _ALL_ the data
        #
        # here, you need to extract the desired artist id!
        #
        return data["results"][0]["artistId"]
        # for the moment, the above just returns all of the data.
    else:
        print(f"the request result was {result}, returning {{}}")  # wow
        return {}    # return no data 

if True:
    #artist_name = "Kendrick Lamar"   # Note:  The code "cheated," sort of...   (See above)
    artist_name = "Taylor Swift"
    # artist_name = "Lizzo"
    # artist_name = "The Beatles"
    result = get_id_from_name(artist_name)
    print(f"artist id for {artist_name} is {result}")
 
```

```python
#
# Your task (#2 of 4)
#
# Write a function     get_albums_from_id(id)
#
# This function should take in an artistId (named id) and return the _whole_ json-obtained dictionary of album data
# 
# For sure, feel free to visit the itunes documentation (linked in the hw).
#
# It's also possible simply to adapt the first function, above, using this url:
#    https://itunes.apple.com/lookup?entity=album&id=472949623&limit=200
#
# If you paste it into your browser, it should either provide the results in a tab -- or directly download them
#    from which you can open them up in vscode :-)
#


def get_albums_from_id(id):
    """ This function takes in 
           id:  a string representing an artist's #id
           and returns the json-obtained dictionary of album data

        Return -1 to signal an error if there was no match.
    """
    # good to keep in mind: you don't _have_ to re-download data, if you have already

    
    search_url = "https://itunes.apple.com/lookup"

    parameters = {  "id":id,
                    "entity":"album",
                    "media":"music",
                    "limit":200  }

    result = requests.get(search_url, params=parameters)
    # print(result.url)
    # print(f"result.status_code is {result.status_code}")
    if result.status_code == 200:
        data = result.json()   # this is _ALL_ the data
        #print(data)
        #
        # here, you need to extract the desired artist id!
        #
        return data
        # for the moment, the above just returns all of the data.
    else:
        print(f"the request result was {result}, returning {{}}")  # wow
        return {}    # return no data 

if True:
    #artist_name = "Kendrick Lamar"   # Note:  The code "cheated," sort of...   (See above)
    artist_name = "Taylor Swift"
    #artist_name = "Lizzo"
    #artist_name = "The Beatles"
    get_albums_from_id(get_id_from_name(artist_name))
   
 
```

```python
#
# Your task (#3 of 4)
#
# Write a function     more_productive( artist_name1, artist_name2 )
#
# This function should 
#    + take two artist names
#    + look up their ids                         [web-access number one]
#    + using those ids, obtain their album data  [web-access number two]
#    + then, from that data, determine who has produced more albums and report that result
#

# here are two example outputs (feel free to format as you'd like...)

def more_productive( artist_name1, artist_name2 ):
    """ This function takes in two artists' names
        Determines how many albums each artist produced
        Then reports who has produced more albums.
    """
    artist_id1 = get_id_from_name(artist_name1)
    artist_id2 = get_id_from_name(artist_name2)

    artist_albums1 = get_albums_from_id(artist_id1)
    artist_albums2 = get_albums_from_id(artist_id2)

    number_albums1 = artist_albums1["resultCount"] -1 #The first result is the artist not a collection
    number_albums2 = artist_albums2["resultCount"] -1

    if  number_albums1 > number_albums2:
        print(f"{artist_name1} produced {number_albums1} albums and {artist_name2} produced {number_albums2} albums.")
        print(f"So {artist_name1} is more productive!")
    elif number_albums1 < number_albums2:
        print(f"{artist_name1} produced {number_albums1} albums and {artist_name2} produced {number_albums2} albums.")
        print(f"So {artist_name2} is more productive!")
    else:
        print(f"{artist_name1} produced {number_albums1} albums and {artist_name2} produced {number_albums2} albums.")
        print("So there is no winner!")



    

"""
*** Comparing Artist Productivity ***

Steve Perry: 11 results
Katy Perry: 87 results
(((Katy Perry has more!)))

Adele: 11 results
Lorde: 17 results
(((Lorde has more!)))
"""

if True:
    katy_perry = "Katy Perry"
    steve_perry = "Steve Perry"
    more_productive(katy_perry, steve_perry)

    taylor_swift = "Taylor Swift"
    more_productive(katy_perry, taylor_swift)

    bts = "BTS"
    more_productive(katy_perry, bts)
```

For fun I decided to look at how many of the albums had the word "Love" in them since a lot of pop songs seem to do with love.

```python
def love( artist_name1, artist_name2 ):
    """ This function takes in two artists' names
        Determines how many albums with the word "Love" each artist produced
        Then reports who has produced more "lovely" albums.
    """
    artist_id1 = get_id_from_name(artist_name1)
    artist_id2 = get_id_from_name(artist_name2)

    artist_albums1 = get_albums_from_id(artist_id1)
    artist_albums2 = get_albums_from_id(artist_id2)

    albums1 = artist_albums1["results"]
    albums2 = artist_albums2["results"]

    love1 = 0
    love2 = 0

    # print(albums1)
    # print(albums2)

    for album in albums1 :
        if album["wrapperType"] == "collection":
            if "Love " in album["collectionName"]:
                love1+=1

    for album in albums2 :
        if album["wrapperType"] == "collection":
            if "Love " in album["collectionName"]:
                love2+=1


    if  love1 > love2:
        print(f"{artist_name1} produced {love1} lovely albums and {artist_name2} produced {love2} lovely albums.")
        print(f"So {artist_name1} is more lovely!")
    elif love1 < love2:
        print(f"{artist_name1} produced {love1} lovely albums and {artist_name2} produced {love2} lovely albums.")
        print(f"So {artist_name2} is more lovely!")
    else:
        print(f"{artist_name1} produced {love1} lovely albums and {artist_name2} produced {love2} lovely albums.")
        print("So there is no winner!")



if True:
    katy_perry = "Katy Perry"
    steve_perry = "Steve Perry"
    love(katy_perry, steve_perry)

    taylor_swift = "Taylor Swift"
    love(katy_perry, taylor_swift)

    bts = "BTS"
    love(katy_perry, bts)
```

For fun I decided to look at how many tracks each artist produced in their albums to see if that changed the winners.

```python
def tracks( artist_name1, artist_name2 ):
    """ This function takes in two artists' names
        Determines how many tracks the artist produced in albums
        Then reports who has produced more tracks in albums.
    """
    artist_id1 = get_id_from_name(artist_name1)
    artist_id2 = get_id_from_name(artist_name2)

    artist_albums1 = get_albums_from_id(artist_id1)
    artist_albums2 = get_albums_from_id(artist_id2)

    albums1 = artist_albums1["results"]
    albums2 = artist_albums2["results"]

    tracks1 = 0
    tracks2 = 0

    # print(albums1)
    # print(albums2)

    for album in albums1 :
        if album["wrapperType"] == "collection":
            tracks1 += int(album["trackCount"])

    for album in albums2 :
        if album["wrapperType"] == "collection":
            tracks2 += int(album["trackCount"])


    if  tracks1 > tracks2:
        print(f"{artist_name1} produced {tracks1} tracks and {artist_name2} produced {tracks2} tracks.")
        print(f"So {artist_name1} is more tracky!")
    elif tracks1 < tracks2:
        print(f"{artist_name1} produced {tracks1} tracks and {artist_name2} produced {tracks2} tracks.")
        print(f"So {artist_name2} is more tracky!")
    else:
        print(f"{artist_name1} produced {tracks1} tracks and {artist_name2} produced {tracks2} tracks.")
        print("So there is no winner!")



if True:
    katy_perry = "Katy Perry"
    steve_perry = "Steve Perry"
    tracks(katy_perry, steve_perry)

    taylor_swift = "Taylor Swift"
    tracks(katy_perry, taylor_swift)

    bts = "BTS"
    tracks(katy_perry, bts)
```
